```python
import tkinter as tk
from PIL import Image, ImageFilter
import time

window = tk.Tk()
window.title("사진 편집기")

entry1 = tk.Entry(window)
entry1.grid(row = 0, column = 0)

def get_png():
    img = str(entry1.get())
    a = Image.open(img)
    a.show()

button1 = tk.Button(window, text = "경로의 사진 보기기", command = get_png)
button1.grid(row = 0, column = 1)

def black():
    img = str(entry1.get())
    a = Image.open(img)
    a = a.convert("L")
    a.show()

button2 = tk.Button(window, text = "흑백 처리", command = black)
button2.grid(row = 1, column = 0)

def blur():
    img = entry1.get()
    a = Image.open(img)
    a = a.filter(ImageFilter.BLUR)
    a.show()

button3 = tk.Button(window, text = "블러 처리", command = blur)
button3.grid(row = 2, column = 0)

def contour():
    img = entry1.get()
    a = Image.open(img)
    a = a.filter(ImageFilter.CONTOUR)
    a.show()

button4 = tk.Button(window, text = "윤곽 처리", command = contour)
button4.grid(row = 3, column = 0)

def emboss():
    img = entry1.get()
    a = Image.open(img)
    a = a.filter(ImageFilter.EMBOSS)
    a.show()

button5 = tk.Button(window, text = "엠보싱", command = emboss)
button5.grid(row = 4, column = 0)

def smooth():
    img = entry1.get()
    a = Image.open(img)
    a = a.filter(ImageFilter.SMOOTH)
    a.show()

button6 = tk.Button(window , text = "부드럽게", command = smooth)
button6.grid(row = 5, column = 0)

def sharpen():
    img = entry1.get()
    a = Image.open(img)
    a = a.filter(ImageFilter.SHARPEN)
    a.show()

button7 = tk.Button(window, text = "선명하게", command = sharpen)
button7.grid(row = 6, column = 0)

window.mainloop()




```

```python
import tkinter as tk
from tkinter import ttk
from PIL import Image, ImageFilter
import shutil
import os

window = tk.Tk()
window.title("사진 편집기")
window.geometry("600x400")

style = ttk.Style()
style.configure("TButton", font=("Helvetica", 12), padding=10)
style.configure("TEntry", font=("Helvetica", 12), padding=10)

entry1 = ttk.Entry(window, width=40)
entry1.grid(row = 0, column = 0, padx=10, pady=10)

backup_image_path = "backup_image.png"

def load_image():
    img_path = str(entry1.get())
    shutil.copy(img_path, backup_image_path)  
    img = Image.open(img_path)
    img.show()

button_load = ttk.Button(window, text = "사진 불러오기(경로를 입력해)", command = load_image)
button_load.grid(row = 0, column = 1, padx=10, pady=10)

def apply_grayscale():
    img_path = str(entry1.get())
    img = Image.open(img_path)
    img = img.convert("L")
    img.save(img_path)
    img.show()

button_grayscale = ttk.Button(window, text = "흑백 처리", command = apply_grayscale)
button_grayscale.grid(row = 1, column = 0, padx=10, pady=10)

def apply_blur():
    img_path = str(entry1.get())
    img = Image.open(img_path)
    img = img.filter(ImageFilter.BLUR)
    img.save(img_path)
    img.show()

button_blur = ttk.Button(window, text = "블러 처리", command = apply_blur)
button_blur.grid(row = 1, column = 1, padx=10, pady=10)

def apply_contour():
    img_path = str(entry1.get())
    img = Image.open(img_path)
    img = img.filter(ImageFilter.CONTOUR)
    img.save(img_path)
    img.show()

button_contour = ttk.Button(window, text = "윤곽 처리", command = apply_contour)
button_contour.grid(row = 2, column = 0, padx=10, pady=10)

def apply_emboss():
    img_path = str(entry1.get())
    img = Image.open(img_path)
    img = img.filter(ImageFilter.EMBOSS)
    img.save(img_path)
    img.show()

button_emboss = ttk.Button(window, text = "엠보싱", command = apply_emboss)
button_emboss.grid(row = 2, column = 1, padx=10, pady=10)

def apply_smooth():
    img_path = str(entry1.get())
    img = Image.open(img_path)
    img = img.filter(ImageFilter.SMOOTH)
    img.save(img_path)
    img.show()

button_smooth = ttk.Button(window, text = "부드럽게", command = apply_smooth)
button_smooth.grid(row = 3, column = 0, padx=10, pady=10)

def apply_sharpen():
    img_path = str(entry1.get())
    img = Image.open(img_path)
    img = img.filter(ImageFilter.SHARPEN)
    img.save(img_path)
    img.show()

button_sharpen = ttk.Button(window, text = "선명하게", command = apply_sharpen)
button_sharpen.grid(row = 3, column = 1, padx=10, pady=10)


window.mainloop()
```
