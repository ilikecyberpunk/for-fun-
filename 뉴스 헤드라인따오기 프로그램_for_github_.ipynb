```python
# BETTER VERSION !!!!!!!!

import tkinter as tk
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup as bs
import requests
import webbrowser

# 색상 설정
bg_color = 'white'  # 배경색
text_color = 'black'  # 텍스트 색상
font_large = ('Helvetica', 16)

window = tk.Tk()
window.geometry('900x600')
window.configure(bg=bg_color)
window.title("뉴스 헤드라인 따오기 !!")

rafio_var = tk.StringVar(value="Google News")

radio1 = tk.Radiobutton(window, text="Google News", variable=rafio_var, value="Google News", bg=bg_color, fg=text_color, selectcolor=bg_color, font=font_large)
radio1.grid(row=0, column=0, padx=20, pady=10, sticky='w')
radio2 = tk.Radiobutton(window, text="Naver News", variable=rafio_var, value="Naver News", bg=bg_color, fg=text_color, selectcolor=bg_color, font=font_large)
radio2.grid(row=0, column=1, padx=20, pady=10, sticky='w')

a = tk.Label(window, text="관심있는 뉴스 키워드 입력 ", bg=bg_color, fg=text_color, font=font_large)
a.grid(row=1, column=0, columnspan=2, padx=20, pady=10, sticky='w')
a1 = tk.Entry(window, bg='white', fg=text_color, insertbackground=text_color, font=font_large)
a1.grid(row=2, column=0, columnspan=2, padx=20, pady=10, sticky='w')

b1 = tk.Text(window, bg='white', fg=text_color, font=font_large, wrap='word', height=15, width=80)
b1.grid(row=4, column=0, columnspan=3, padx=20, pady=10, sticky='w')

def search():
    b1.delete(1.0, tk.END)
    if rafio_var.get() == "Naver News":
        url = 'https://search.naver.com/search.naver?where=news&sm=tab_jum&query=' + '%s' % a1.get()
        r = requests.get(url)
        html_text = r.text

        soup = bs(r.text, 'html.parser')
        news_titles = soup.select("a.news_tit")
        for i in news_titles:
            title = i.get_text()
            href = i.attrs['href']
            b1.insert(tk.END, f"{title}\n{href}\n\n")

    elif rafio_var.get() == "Google News":
        url = 'https://www.google.com/search?q=' + a1.get() + '&tbm=nws'
        r = requests.get(url)
        html_text = r.text

        soup = bs(html_text, 'html.parser')
        news_titles = soup.select("div.BNeawe.vvjwJb.AP7Wnd")
        for i in news_titles:
            title = i.get_text()
            parent = i.find_parent('a')
            href = parent.attrs['href']
            b1.insert(tk.END, f"{title}\nhttps://www.google.com{href}\n\n")

def d():
    b1.delete(1.0, tk.END)

def save_to_file():
    with open("news_results.txt", "w", encoding="utf-8") as file:
        file.write(b1.get(1.0, tk.END))

def open_in_browser(event):
    index = b1.index("@%s,%s" % (event.x, event.y))
    line = b1.get(index + " linestart", index + " lineend")
    if line.startswith("http"):
        webbrowser.open(line)

a2 = tk.Button(window, text="검색 !!", command=search, bg=bg_color, fg=text_color, font=font_large)
a2.grid(row=3, column=0, padx=20, pady=10, sticky='w')
b3 = tk.Button(window, text="삭제 !!", command=d, bg=bg_color, fg=text_color, font=font_large)
b3.grid(row=3, column=1, padx=20, pady=10, sticky='w')
b4 = tk.Button(window, text="저장 !!", command=save_to_file, bg=bg_color, fg=text_color, font=font_large)
b4.grid(row=3, column=2, padx=20, pady=10, sticky='w')

b1.bind("<Button-1>", open_in_browser)

window.mainloop()
```
