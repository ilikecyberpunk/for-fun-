```python
# rsa 공개키 암호 프로그램 
import tkinter as tk
from tkinter import messagebox

root = tk.Tk()
root.title("rsa 공개키 암호화")

tk.Label(root, text="소수 p를 입력하세요").grid(row=0, column=0)
entryP = tk.Entry(root)
entryP.grid(row=0, column=1)

tk.Label(root, text="소수 q를 입력하세요").grid(row=1, column=0)
entryQ = tk.Entry(root)
entryQ.grid(row=1, column=1)

def Euclidean_algorithm(x, y):
    a, b = max(x, y), min(x, y)  
    while b != 0: 
        r = a % b  
        a, b = b, r  
    return a  


def is_prime(n):
    n = int(n)  
    if n < 2:  
        return False
    for i in range(2, int(n**0.5) + 1):  
        if n % i == 0: 
            return False
    return True  



def create_key():
    try:
        p = int(entryP.get())
        q = int(entryQ.get())

        if is_prime(p) and is_prime(q):
            global z, n, s
            z = p * q
            phi_z = (p - 1) * (q - 1)
            
            # gcd(n, phi_z) == 1인 정수 
            n = 2
            while n < phi_z and Euclidean_algorithm(n, phi_z) != 1:
                n += 1

            # n*s mod phi_z == 1을 만족하는 정수
            s = 1
            while (n * s) % phi_z != 1:
                s += 1
            
            label_public_key.config(text=f"공개키 : ({n}, {z})")
            label_private_key.config(text=f"개인키 : ({s}, {z})")
        else:
            messagebox.showerror("오류", "p와 q에 소수를 입력하세요")
    except Exception as e:
        messagebox.showerror("오류", f"{e}")



tk.Button(root, text="키생성!", command=create_key).grid(row=2, column=0)

#==========================================================================================================

label_public_key = tk.Label(root, text="")
label_public_key.grid(row=3, column=0)

label_private_key = tk.Label(root, text="")
label_private_key.grid(row=4, column=0)

tk.Label(root, text="암호화할 메세지(10진수)를 입력하세요").grid(row=7, column=0)
entry_message = tk.Entry(root)
entry_message.grid(row=7, column=1)



def XD():
    try:
        global z, n
        message = int(entry_message.get())  

        
        encrypted_message = pow(message, n, z)  

        label_message.config(text=f"암호화된 메세지 : {encrypted_message}")

    except Exception as e:
        messagebox.showerror("오류", f"{e}")


tk.Button(root, text="암호화 !!", command=XD).grid(row=8, column=0)


label_message = tk.Label(root, text="")
label_message.grid(row=9, column=0)
#------------------------------------------------------------------------------------------------------------

tk.Label(root, text = "복호화할 메세지를 입력하세요(10진수)").grid(row = 11, column = 0)
entry_decryation = tk.Entry(root)
entry_decryation.grid(row = 11, column = 1)

def decryation():
    try:
        global s, z

        d_message = int(entry_decryation.get())

        decrypted_message = pow(d_message, s, z)

        label_d_message.config(text = f"복호화된 메세지 : {decrypted_message}")
    
    except Exception as e:
        messagebox.showerror("오류", f"{e}")


tk.Button(root, text = "복호화 !!", command=decryation).grid(row = 12, column = 0)

label_d_message = tk.Label(root, text = "")
label_d_message.grid(row = 13, column = 0)






root.mainloop()
```
