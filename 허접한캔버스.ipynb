```python
import tkinter as tk
from tkinter import messagebox
import pyautogui

root = tk.Tk()
root.title("minicanvas")

#도구창
frame1 = tk.LabelFrame(root, text = "Tools", font = ("Times New Roman", 14, "bold"))
frame1.grid(row= 0, column = 0)

# 네모 생성
frame_of_rec = tk.LabelFrame(frame1, text = "네모", font = ("Times New Roman", 10, "bold"))
frame_of_rec.grid(row = 0, column = 0)

label_rec = tk.Label(frame_of_rec, text = "높이").grid(row = 0, column = 0)
entry_rec = tk.Entry(frame_of_rec)
entry_rec.grid(row = 0, column = 1)

label_rec2 = tk.Label(frame_of_rec, text = "밑변 길이").grid(row = 1, column = 0)
entry_rec2 = tk.Entry(frame_of_rec)
entry_rec2.grid(row = 1, column = 1)

location_of_rec = tk.Label(frame_of_rec, text = "위치 설정").grid(row = 2, column = 0)
x_of_rec = tk.Label(frame_of_rec, text = "x = ").grid(row = 3, column = 0)
entry_x = tk.Entry(frame_of_rec)
entry_x.grid(row = 3, column = 1)
y_of_rec = tk.Label(frame_of_rec, text = "y = ").grid(row = 4, column = 0)
entry_y = tk.Entry(frame_of_rec)
entry_y.grid(row = 4, column = 1)

def generate_rec():
    try:
        width = int(entry_rec2.get())
        height = int(entry_rec.get())
        x1, y1 = int(entry_x.get()), int(entry_y.get())
        x2, y2 = x1 + width, y1 + height
        canvas.create_rectangle(x1, y1, x2, y2, tags = "rect")
    except Exception as e:
        messagebox.showerror("오류!!", f"{e}")

button_rec = tk.Button(frame_of_rec, text = "생성", font = ("Times New Roman", 10, "bold"), command=generate_rec)
button_rec.grid(row = 5, column=0)

def delete_rect():
    try:
        canvas.delete("rect")
    except Exception as e:
        messagebox.showerror("오류", f"{e}")

button_rec_delete = tk.Button(frame_of_rec, text = "네모 삭제", font = ("Times New Roman", 10, "bold"), command = delete_rect).grid(row = 5, column = 1)

#동그라미
frame_of_circle = tk.LabelFrame(frame1, text = "동그라미", font = ("Times New Roman", 10, "bold"))
frame_of_circle.grid(row = 2, column = 0)

label_start_x = tk.Label(frame_of_circle, text = "위치설정 x = ").grid(row = 0, column = 0)
entry_start_x = tk.Entry(frame_of_circle)
entry_start_x.grid(row = 0, column = 1)

label_start_y = tk.Label(frame_of_circle, text = "위치설정 y = ").grid(row = 1, column = 0)
entry_start_y = tk.Entry(frame_of_circle)
entry_start_y.grid(row = 1, column = 1)

label_width_of_circle = tk.Label(frame_of_circle, text = "지름").grid(row = 2, column = 0)
entry_width_of_circle = tk.Entry(frame_of_circle)
entry_width_of_circle.grid(row = 2, column = 1)

label_height_of_circle = tk.Label(frame_of_circle,text = "높이").grid(row = 3, column = 0)
entry_height_of_circle = tk.Entry(frame_of_circle)
entry_height_of_circle.grid(row = 3, column = 1)

def generate_circle():
    try:
        x = int(entry_start_x.get())
        y = int(entry_start_y.get())

        width = int(entry_width_of_circle.get())
        height = int(entry_height_of_circle.get())

        x2 = x + width
        y2 = y + height

        canvas.create_oval(x, y, x2, y2)
    
    except Exception as e:
        messagebox.showerror("오류", f"{e}")
    

button_of_circle = tk.Button(frame_of_circle, text = "생성", command = generate_circle, font = ("Times New Roman", 10, "bold")).grid(row = 4, column = 0)

def delete_circle():
    try:
        canvas.delete("circle")
    except Exception as e:
        messagebox.showerror("오류", f"{e}")

button_of_del_circle = tk.Button(frame_of_circle, text = "동그라미 삭제", font = ("Times New Roman", 10, "bold"), command = delete_circle).grid(row  =4, column = 1)

#세모
frame_of_triangle = tk.LabelFrame(frame1, text="삼각형", font=("Times New Roman", 10, "bold"))
frame_of_triangle.grid(row=3, column=0)

tk.Label(frame_of_triangle, text="x1").grid(row=0, column=0)
entry_tri_x1 = tk.Entry(frame_of_triangle)
entry_tri_x1.grid(row=0, column=1)

tk.Label(frame_of_triangle, text="y1").grid(row=1, column=0)
entry_tri_y1 = tk.Entry(frame_of_triangle)
entry_tri_y1.grid(row=1, column=1)

tk.Label(frame_of_triangle, text="x2").grid(row=2, column=0)
entry_tri_x2 = tk.Entry(frame_of_triangle)
entry_tri_x2.grid(row=2, column=1)

tk.Label(frame_of_triangle, text="y2").grid(row=3, column=0)
entry_tri_y2 = tk.Entry(frame_of_triangle)
entry_tri_y2.grid(row=3, column=1)

tk.Label(frame_of_triangle, text="x3").grid(row=4, column=0)
entry_tri_x3 = tk.Entry(frame_of_triangle)
entry_tri_x3.grid(row=4, column=1)

tk.Label(frame_of_triangle, text="y3").grid(row=5, column=0)
entry_tri_y3 = tk.Entry(frame_of_triangle)
entry_tri_y3.grid(row=5, column=1)

def generate_triangle():
    try:
        x1 = int(entry_tri_x1.get())
        y1 = int(entry_tri_y1.get())

        x2 = int(entry_tri_x2.get())
        y2 = int(entry_tri_y2.get())

        x3 = int(entry_tri_x3.get())
        y3 = int(entry_tri_y3.get())

        canvas.create_polygon(x1, y1, x2, y2, x3, y3, fill="", outline="black", width=2, tags="triangle")

    except Exception as e:
        messagebox.showerror("오류", f"{e}")

button_of_triangle = tk.Button(frame_of_triangle, text="생성", command=generate_triangle, font=("Times New Roman", 10, "bold"))
button_of_triangle.grid(row=6, column=0)

def delete_triangle():
    try:
        canvas.delete("triangle")
    except Exception as e:
        messagebox.showerror("오류", f"{e}")

button_of_del_triangle = tk.Button(frame_of_triangle, text="세모 삭제", command=delete_triangle, font=("Times New Roman", 10, "bold"))
button_of_del_triangle.grid(row=6, column=1)

# 캔버스
frame2 = tk.LabelFrame(root, text = "!~~~ HuhJupHanCanvas ~~~!\n펜으로 그리시려면 마우스를 왼클릭하시고 드래그하세요 (천천히 부드럽게)", font = ("Times New Roman", 12, "bold"))
frame2.grid(row = 0, column = 1)

canvas = tk.Canvas(frame2, width= 1000, height = 700, bg = "white")
canvas.grid(row = 0, column = 0)

#전체삭제버튼
def delete():
    canvas.delete(tk.ALL)

button_of_del = tk.Button(frame2, text = "전체삭제", command = delete)
button_of_del.grid(row = 1, column = 0)

#펜
def draw_circle(event):
    x, y = event.x, event.y
    canvas.create_oval(x - 1, y - 1, x + 1, y + 1)
canvas.bind("<B1-Motion>", draw_circle)


root.mainloop()
```
